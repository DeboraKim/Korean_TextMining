```
def get_KRWordRank(morphs):
    from krwordrank.word import KRWordRank

    min_count = 5
    max_length = 20
    wordrank_extractor = KRWordRank(min_count = min_count, max_length = max_length)

    beta = 0.85 
    max_iter = 10
    texts = morphs
    keywords, rank, grapy = wordrank_extractor.extract(texts, beta, max_iter)

    '''for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True)[:100]:
      print("%s: %.4f" %(word, r))
    '''
    return keywords
  
def get_hannanum(word_list):
  from konlpy.tag import Hannanum
  hannanum = Hannanum()

  words = hannanum.pos(word_list)

  output = []

  for word, tag in words:
      if tag in ['H','N']:
        output.append(word)

  return output

file = open('/content/drive/My Drive/Colab Notebooks/패션아이템_원문/TOP_ZIPUP_2019.txt', 'r')

lists = file.readlines()

file.close()

morphs = []

for list in lists:
    morphs.append(list)

keywords = get_KRWordRank(morphs)

word_list =[]
for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True):
    word_list.append(word)
word_list = " ".join(word_list)

output = []
output = get_hannanum(word_list)


with open('/content/drive/My Drive/Colab Notebooks/output_hannanum.txt', 'w') as f:
    for line in output:
        f.write("\n"+line)

  
```
